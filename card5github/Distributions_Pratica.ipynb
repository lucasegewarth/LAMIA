```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

values = np.random.uniform(-40.0, 40.0, 100000) #gera uma distribuição de dados nos caso 10000 de -40 a 40 e no caso é aletoria e uniforme
plt.hist(values, 50)
plt.show()
```

```python
from scipy.stats import norm # fornece funções para trabalhar com distribuições normais
import matplotlib.pyplot as plt

x = np.arange(-6, 6, 0.001)# são valores entre -6 e 6, e no total 12000 [-6.000, -5.999, -5.998, ..., 5.998, 5.999, 6.000]
plt.plot(x, norm.pdf(x)) # norm é usado é usado para plotar os valores e como visto no gráfico eles estão mais proximos de 0
```

```python
import numpy as np
import matplotlib.pyplot as plt

media = 12.0 # Media
DesvioPadrao = 3.5 # Desvio padrão
values = np.random.normal(media, DesvioPadrao, 100000) #distribuição aletoria normal
plt.hist(values, 50)
plt.show()
```

```python
from scipy.stats import expon #  Fornece funções para trabalhar com a distribuição exponencial
import matplotlib.pyplot as plt # Para plotar o gráfico

x = np.arange(0, 10, 0.001) # Gera valores de 0 a 10, com incrementos de 0.001
plt.plot(x, expon.pdf(x)) #  expon.pdf(x) = Calcula a função de densidade de probabilidade (PDF) para cada valor de X
```

```python
from scipy.stats import binom #Fornece funções para trabalhar com a distribuição binomial
import matplotlib.pyplot as plt

n, p = 10, 0.5 # Número de tentativas no experimento e Probabilidade de sucesso em cada tentativa
x = np.arange(0, n+1, 1)
plt.plot(x, binom.pmf(x, n, p))
```

```python
from scipy.stats import poisson # Classe do SciPy que modela a distribuição de Poisson
import matplotlib.pyplot as plt

mu = 1000 # representa a média esperada de ocorrências por intervalo de tempo ou espaço
x = np.arange(800, 1200, 0.5)
plt.plot(x, poisson.pmf(x, mu))
```
