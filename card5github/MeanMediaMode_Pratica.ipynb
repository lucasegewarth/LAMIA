```python
%matplotlib inline
import numpy as np # é uma biblioteca para computação numérica
import matplotlib.pyplot as plt # é usada para criar gráficos e visualizações

incomes = np.random.normal(150.0, 35.0, 100000) #primeiro numero se refere a média, o segundo seria o desvio padrão, e terceiro a quantidade total de valores

plt.hist(incomes, 100) #plota um histograma com 100 intervalor o 100 "barinhas"
plt.show() # exibe ohistograma 
```

```python
np.median(incomes) # peço a mediana dos valores
```

Obsevace que o valor não exatos 150 mas um valor muito proximo do mesmo

```python
np.mean(incomes) # peço a média do meu gráfico 
```

```python
idades = np.random.randint(18, high=31, size=1000)
idades
```

```python
plt.hist(idades, bins=40, edgecolor='black')
plt.xlabel('Idades')
plt.ylabel('Frequência')
plt.title('Distribuição de Idades da UTFPR-SH')
plt.show()
```

```python
plt.hist(idades, bins=range(min(idades), max(idades) + 2), edgecolor='black', align='left')
plt.xlabel('Idades')
plt.ylabel('Frequência')
plt.title('Distribuição de Idades da UTFPR-SH')
plt.xticks(range(min(idades), max(idades) + 1))  # Define os valores do eixo X de 1 em 1
plt.show()
```
