```python
x = np.arange(-6, 6, 0.01)

fig, axes = plt.subplots()  # Forma mais segura de criar os eixos

axes.set_xlim([-5, 5])
axes.set_ylim([0, 1.0])
axes.set_xticks(np.arange(-5, 6, 1))  # Gera os ticks automaticamente de -5 a 5
axes.set_yticks(np.arange(0, 1.1, 0.1))  # Gera os ticks de 0 a 1.0

# Plotando as curvas
axes.plot(x, norm.pdf(x), label="N(0,1)", color="blue")
axes.plot(x, norm.pdf(x, 1.0, 0.5), label="N(1.0,0.5)", color="red", linestyle="dashed")

axes.legend()
axes.grid(True)  # Adiciona uma grade para facilitar a leitura
axes.set_title("Distribuies Normais")
axes.set_xlabel("X")
axes.set_ylabel("Densidade de Probabilidade")

plt.show()
```

```python
# Criando os dados
x = np.arange(-6, 6, 0.01)

# Criando a figura e os eixos
fig, axes = plt.subplots(figsize=(7, 5))  # Melhor controle do tamanho da figura

# Configurando os limites e ticks
axes.set_xlim([-5, 5])
axes.set_ylim([0, 1.0])
axes.set_xticks(np.arange(-5, 6, 1))  # Ticks no eixo X de -5 a 5
axes.set_yticks(np.arange(0, 1.1, 0.1))  # Ticks no eixo Y de 0 a 1.0

# Adicionando grid
axes.grid(True)

# Definindo rtulos personalizados dos eixos
plt.xlabel("Greebles")  
plt.ylabel("Probability")  

# Plotando as curvas com estilos e cores
plt.plot(x, norm.pdf(x), 'b-', label="Sneetches")  # Linha azul contnua
plt.plot(x, norm.pdf(x, 1.0, 0.5), 'r--', label="Gacks")  # Linha vermelha tracejada

# Adicionando legenda no canto inferior direito
plt.legend(loc=4)  

# Exibindo o grfico
plt.show()
```

```python
# Gerando os dados aleatrios
X = np.random.randn(500)
Y = np.random.randn(500)

# Criando o grfico de disperso com melhorias
plt.figure(figsize=(6,6))  # Ajusta o tamanho da figura
plt.scatter(X, Y, color='blue', alpha=0.5, edgecolor='black')  # Adiciona transparncia e borda
plt.axhline(0, color='black', linestyle='dashed', linewidth=1)  # Linha horizontal no 0
plt.axvline(0, color='black', linestyle='dashed', linewidth=1)  # Linha vertical no 0
plt.grid(True, linestyle="--", alpha=0.6)  # Adiciona grade pontilhada
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.title("Grfico de Disperso de uma Distribuio Normal")

plt.show()
```

famoso grfico de bigodes ou boxplot

```python
uniformSkewed = np.random.rand(100) * 100 - 40
high_outliers = np.random.rand(10) * 50 + 100
low_outliers = np.random.rand(10) * -50 - 100
data = np.concatenate((uniformSkewed, high_outliers, low_outliers))
plt.boxplot(data)
plt.show()
```

```python
# Gerando os dados
uniformSkewed = np.random.rand(100) * 100 - 40
high_outliers = np.random.rand(10) * 50 + 100
low_outliers = np.random.rand(10) * -50 - 100
data = np.concatenate((uniformSkewed, high_outliers, low_outliers))

# Criando o boxplot com melhorias
plt.figure(figsize=(6, 5))  # Ajusta o tamanho da figura
plt.boxplot(data, patch_artist=True, 
            boxprops=dict(facecolor="lightblue"),  # Cor da caixa
            medianprops=dict(color="red", linewidth=2),  # Linha da mediana
            flierprops=dict(marker='o', color='orange', markersize=8, alpha=0.6))  # Outliers

# Configuraes do grfico
plt.title("Boxplot de Dados com Outliers", fontsize=14)
plt.ylabel("Valores", fontsize=12)
plt.grid(axis='y', linestyle="--", alpha=0.7)  # Grade no eixo Y

# Exibindo o grfico
plt.show()
```

Tentei criar um grfico de disperso representando dados aleatrios sobre idade versus tempo gasto assistindo TV e com cores

```python
# Gerando os dados
X = np.random.normal(loc=25, scale=25, size=500)  # Idade com mdia 25 anos e desvio padro 25 anos
Y = np.random.uniform(0, 24, 500)  # Tempo aleatrio entre 0 e 24 horas
X = np.clip(X, 3, 80)  # Limitando as idades entre 3 e 80 anos

# Criando o scatter plot
plt.figure(figsize=(8, 5))  # Ajusta o tamanho do grfico
scatter = plt.scatter(X, Y, c=X, cmap='coolwarm', alpha=0.7, edgecolor='black')  # Cores baseadas na idade

# Adicionando ttulo, rtulos e configuraes extras
plt.title("Grfico de Disperso: Idade vs. Tempo", fontsize=14)
plt.xlabel("Idade", fontsize=12)
plt.ylabel("Tempo (horas)", fontsize=12)
plt.xlim(0, 80)
plt.ylim(0, 24)
plt.grid(True, linestyle="--", alpha=0.6)  # Grade pontilhada
plt.colorbar(scatter, label="Idade")  # Barra de cores indicando a idade

# Exibindo o grfico
plt.show()
```
