```python
import numpy as np
import matplotlib.pyplot as plt

# Funes de covarincia e correlao
def de_mean(x):
    return x - np.mean(x)

def covariance(x, y):
    return np.dot(de_mean(x), de_mean(y)) / (len(x) - 1)

def correlation(x, y):
    stddevx = x.std()
    stddevy = y.std()
    
    if stddevx == 0 or stddevy == 0:
        return 0  

    return covariance(x, y) / (stddevx * stddevy)

# Gerando trs conjuntos de dados
np.random.seed(42)  # Para reprodutibilidade

studyHours = np.random.normal(5, 2, 1000)  # Horas de estudo (mdia 5h, desvio 2h)
studyHours = np.clip(studyHours, 0, 10)  # Garantir que no tenha valores negativos

# 1. Dados aleatrios sem relao
examScores1 = np.random.normal(70, 15, 1000)  # Notas sem relao com as horas

# 2. Relao linear positiva
examScores2 = 50 + studyHours * 5  # Notas aumentam linearmente com as horas

# 3. Relao no linear (retornos decrescentes)
examScores3 = 50 + np.log(1 + studyHours) * 30  # Curva logartmica

# Criando subplots
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Grfico 1: Disperso sem relao
axs[0].scatter(studyHours, examScores1)
axs[0].set_title("Sem Relao")
axs[0].set_xlabel("Horas de Estudo")
axs[0].set_ylabel("Nota na Prova")

# Grfico 2: Relao Linear Positiva
axs[1].scatter(studyHours, examScores2)
axs[1].set_title("Relao Linear Positiva")
axs[1].set_xlabel("Horas de Estudo")
axs[1].set_ylabel("Nota na Prova")

# Grfico 3: Relao No Linear
axs[2].scatter(studyHours, examScores3)
axs[2].set_title("Relao No Linear (Retornos Decrescentes)")
axs[2].set_xlabel("Horas de Estudo")
axs[2].set_ylabel("Nota na Prova")

# Ajustando layout
plt.tight_layout()
plt.show()
```

```python
# Funes de covarincia e correlao
def de_mean(x):
    return x - np.mean(x)

def covariance(x, y):
    return np.dot(de_mean(x), de_mean(y)) / (len(x) - 1)

def correlation(x, y):
    stddevx = x.std()
    stddevy = y.std()
    
    if stddevx == 0 or stddevy == 0:
        return 0  

    return covariance(x, y) / (stddevx * stddevy)

# Gerando trs conjuntos de dados
np.random.seed(42)  # Para reprodutibilidade

temperature = np.random.normal(30, 5, 1000)  # Temperatura mdia de 30C, desvio de 5C
temperature = np.clip(temperature, 10, 45)  # Limitando entre 10C e 45C

# 1. Dados aleatrios sem relao
iceCreamSales1 = np.random.normal(200, 50, 1000)  # Vendas aleatrias

# 2. Relao linear positiva
iceCreamSales2 = 50 + temperature * 10  # Vendas aumentam linearmente com a temperatura

# 3. Relao no linear (crescimento desacelerado)
iceCreamSales3 = 50 + np.sqrt(temperature) * 50  # Crescimento mais rpido no incio, depois desacelera

# Criando subplots
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Grfico 1: Disperso sem relao
axs[0].scatter(temperature, iceCreamSales1)
axs[0].set_title("Sem Relao")
axs[0].set_xlabel("Temperatura (C)")
axs[0].set_ylabel("Vendas de Sorvete")

# Grfico 2: Relao Linear Positiva
axs[1].scatter(temperature, iceCreamSales2)
axs[1].set_title("Relao Linear Positiva")
axs[1].set_xlabel("Temperatura (C)")
axs[1].set_ylabel("Vendas de Sorvete")

# Grfico 3: Relao No Linear
axs[2].scatter(temperature, iceCreamSales3)
axs[2].set_title("Relao No Linear (Crescimento Desacelerado)")
axs[2].set_xlabel("Temperatura (C)")
axs[2].set_ylabel("Vendas de Sorvete")

# Ajustando layout
plt.tight_layout()
plt.show()
```
