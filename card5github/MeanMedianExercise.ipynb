```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

incomes = np.random.normal(100.0, 20.0, 10000)

plt.hist(incomes, 50)
plt.show()
```


Agora, encontre a média e a mediana desses dados. No bloco de código abaixo, escreva seu código e veja se o resultado faz sentido:

```python
np.median(incomes)
```

```python
np.mean(incomes)
```
